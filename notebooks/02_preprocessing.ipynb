```python
import pandas as pd
```

```python
df1 = pd.read_csv('../가맹점_현황.csv')
df2 = pd.read_csv('../판매_현황.csv')
df3 = pd.read_csv('../회수_현황.csv')
```

```python
# 결측치 처리
df1 = df1.dropna(subset=['가맹점명'], ignore_index=True)
df1['취급품목'] = df1['취급품목'].fillna('미분류')
```

```python
# 결측치 제거 후 데이터 
df1.shape
```

```python
# 각 컬럼에 어떤 값들이 들어있는지 확인 (이게 이상치 탐색을 대신함)
# 수치적 이상치 제거는 하지 않는다. 대신 데이터의 형식을 맞추는 '표준화'에 집중한다.
df1['소재지'].unique()
```

```python
# 결측치 이상치 이상치 없는 .가맹점_현황.csv 저장
df1.to_csv('../cleaned_onnuri.csv', index=False, encoding='utf-8-sig')
```
